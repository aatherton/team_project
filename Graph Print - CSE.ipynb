# Graph Printer - CSE

```python
import pandas
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

CSV = "actually_final_i_promise.csv"

with open(CSV, "r") as inbox:
    # DATA is the original data
    df = pandas.read_csv(inbox)
college = ['college']
```

```python
df = df.dropna(axis=0, subset=college)

result = {}
for each in df["college"].unique():
    result[each] = {}
    ball = df.loc[df["college"] == each]
    result[each]["playercount"] = len(ball["name"].unique())
    result[each]["average score"] = ball["total annual points"].sum() / result[each]["playercount"]
result = pandas.DataFrame(result).transpose()

df['college'] = (df['college'].astype(str).str.lower())
```

```python
qb_df = pandas.DataFrame(df[df['Position'] == 'QB'])
rb_df = pandas.DataFrame(df[df['Position'] == 'RB'])
te_df = pandas.DataFrame(df[df['Position'] == 'TE'])
wr_df = pandas.DataFrame(df[df['Position'] == 'WR'])
fb_df = pandas.DataFrame(df[df['Position'] == 'FB'])

qb = pandas.DataFrame(qb_df.groupby(['college'])['total annual points'].sum().sort_values(0, ascending=False).nlargest(20))
rb = pandas.DataFrame(rb_df.groupby(['college'])['total annual points'].sum().sort_values(0, ascending=False).nlargest(20))
wr = pandas.DataFrame(wr_df.groupby(['college'])['total annual points'].sum().sort_values(0, ascending=False).nlargest(20))
te = pandas.DataFrame(te_df.groupby(['college'])['total annual points'].sum().sort_values(0, ascending=False).nlargest(20))
fb = pandas.DataFrame(fb_df.groupby(['college'])['total annual points'].sum().sort_values(0, ascending=False).nlargest(20))
```

```python
plt.style.use('seaborn')
qb.index = [x.upper() for x in qb.index]
qb.sort_values(by='total annual points', ascending=True).plot(kind='barh', color='r', legend=None)
plt.title('Top 20 Fantasy Points by College:  QB')
plt.ylabel('College')
plt.xlabel('Total Points: 2000-2016')
plt.xlim(0,5500)
plt.savefig('Top 20 Fantasy Poducing Colleges - QB.png')
plt.show()
```

```python
qb_sum = pandas.DataFrame(qb_df.groupby(['college']).sum())
rb_sum = pandas.DataFrame(rb_df.groupby(['college']).sum())
wr_sum = pandas.DataFrame(wr_df.groupby(['college']).sum())
te_sum = pandas.DataFrame(te_df.groupby(['college']).sum())
fb_sum = pandas.DataFrame(fb_df.groupby(['college']).sum())
```
